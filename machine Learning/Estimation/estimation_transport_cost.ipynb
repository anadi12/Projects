```python
# Assuming lr2 is your original DataFrame and x_train contains the features
# x_train columns should correspond to features in lr2
# reg is your trained LinearRegression model

# Create feature importance DataFrame
feature_importance = pd.DataFrame({
    'Feature': x_train.columns,       # feature names
    'Coefficient': reg.coef_,         # linear regression coefficients
    'Importance': np.abs(reg.coef_)   # absolute value as importance
})

# Sort by importance
feature_importance = feature_importance.sort_values(by='Importance', ascending=False)

print(feature_importance)
```

```python
# Intercept
intercept = reg.intercept_

# Coefficients
coefficients = reg.coef_

# Feature names (use same X_train you fitted on)
features = x_train.columns

# Build regression equation
equation = f"Spend = {intercept:.2f}"
for feat, coef in zip(features, coefficients):
    equation += f" + ({coef:.2f} * {feat})"

print("Linear Regression Equation:")
print(equation)
```

-----Decision Tree----

```python
from sklearn.model_selection import train_test_split
```

```python

from sklearn.tree import DecisionTreeRegressor
```

```python

import sklearn.metrics as metrics
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
dt_model = DecisionTreeRegressor(random_state=42)
```

```python
dt_model.fit(x_train, y_train)
```

```python
y_pred_dt = dt_model.predict(x_test)
```

```python
r2_dt = metrics.r2_score(y_test, y_pred_dt)
mae_dt = metrics.mean_absolute_error(y_test, y_pred_dt)
mse_dt = metrics.mean_squared_error(y_test, y_pred_dt)
```
