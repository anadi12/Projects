```python
dm_cat=dm.select_dtypes(object)
```

```python
dm_num=dm.select_dtypes(int)
```

```python
dummies = pd.get_dummies(dm_cat)
```

```python
dm1=pd.concat([dummies,dm_num],axis="columns")
```

```python
dm1.head()
```

```python
x=dm1.iloc[:,0:18]
```

```python
x
```

```python
y=dm1.iloc[:,18:19]
```

```python
y
```

```python
import sklearn 
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.3,random_state=24)
```

```python
from sklearn.ensemble import GradientBoostingRegressor
```

```python
gb=GradientBoostingRegressor(max_depth=3,n_estimators=120,learning_rate=0.10)
```

```python
gb=gb.fit(xtrain,ytrain)
```

```python
y_pred=gb.predict(xtest)
```

```python
import sklearn.metrics as metrics 
```

```python
mae = metrics.mean_absolute_error(ytest, y_pred)
mse = metrics.mean_squared_error(ytest, y_pred)
```

```python
mae
```

```python
mse
```

```python
m = np.mean(ytest)
```

```python
d = ytest-m
```

```python
mae_base = np.mean(abs(d))
mse_base = np.mean(d**2)
```
